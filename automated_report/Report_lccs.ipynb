```python
#from docx import Document
from docxcompose.composer import Composer
from docx import Document as Document_compose
from docx.shared import Cm
from docxtpl import DocxTemplate, InlineImage
from docx.shared import Cm, Inches, Mm, Emu
import csv 
import os.path
from os import path
```

```python
with open('CAFI.csv', 'r', encoding='UTF-8', newline='') as csv_file:
    data = csv.reader(csv_file,  delimiter=',')
    first_line = True
    rows = []
    image_plot = []
    for row in data:
       
        #print(', '.join(row))
        if not first_line:
            rows.append({
                "code": row[0],
                "group": row[1],
                "nom": row[2],
                "description": row[3],
                "diagram": row[4],
                "xposition": row[5],
                "yposition": row[6],
                "photoplot": row[7],
                "photoplanet": row[8],
                "pays": row[9],
            })
            image_plot = row[7]

        else:
            first_line = False
```

```python
code = []
group = []
nom = []
description = []
xposition = []
yposition = []
pays = []

image_name_diagram = []
image_name_plot = []
image_name_planet = []
files = []

i = 0
for row in rows:
    i = i + 1 
    
    #Import template document
    doc = DocxTemplate('Report_template.docx')
    
   
    code.append(row['code'])
    group.append(row['group'])
    nom.append(row['nom'])
    
    description.append(row['description'])
    xposition.append(row['xposition'])
    yposition.append(row['yposition'])
    pays.append(row['pays'])
    
    image_name_plot.append(row['photoplot'])
    image_name_planet.append(row['photoplanet'])
    image_name_diagram.append(row['diagram'])
    
    if os.path.exists('photo_plot/'+row['photoplot']):
        image1 = InlineImage(doc,'photo_plot/'+row['photoplot'],width=Cm(11), height=Cm(7))
    else:
        print("The file image1 does not exist")
    
    if os.path.exists('photo_planet/'+row['photoplanet']):
        image2 = InlineImage(doc,'photo_planet/'+row['photoplanet'],width=Cm(11), height=Cm(7))
    else:
        print("The file image2 does not exist")
    
    if os.path.exists('lccs_diagram/'+row['diagram']):
        image3 = InlineImage(doc,'lccs_diagram/'+row['diagram'],width=Cm(19), height=Cm(5))
    else:
        print("The file image3 does not exist")
    
    #image1 = InlineImage(doc,'photo_plot/'+row['photoplot'],width=Cm(11), height=Cm(7))
    #image2 = InlineImage(doc,'photo_planet/'+row['photoplanet'],width=Cm(11), height=Cm(7))
    #image3 = InlineImage(doc,'lccs_diagram/'+row['diagram'],width=Cm(19), height=Cm(5))
    
    #print ("File exists:"+str(path.exists('photo_plot/'+row['photoplot'])))

    #print ("Is it File?" + str(path.isfile('photo_plot/'+row['photoplot'])))
    
    #Declare template variables
    context = {
    'title': row['nom'],
    'description' : row['description'],
    'groupe' : row['group'],
    'pays' : row['pays'],
   
    #'table_contents': table_contents,
    'image1': image1,
    'image2': image2,
    'image3': image3
    }
    
    #Render automated report
    doc.render(context)
    #doc.save('generated_report.docx')
    doc.save('generated_report_'+str(i)+'.docx')
    files.append('generated_report_'+str(i)+'.docx')
```

```python
# function to merge all the reports in one file
def combine_all_docx(filename_master,files_list):
    number_of_sections=len(files_list)
    master = Document_compose(filename_master)
    composer = Composer(master)
    for i in range(0, number_of_sections):
        doc_temp = Document_compose(files_list[i])
        composer.append(doc_temp)
    composer.save("Report.docx")
```

```python
filename_master="empty.docx"
```

```python
# merge files 
combine_all_docx(filename_master,files)
```
